```python
import ipywidgets as widgets
```

```python
h = widgets.Dropdown(
    options=video['channel_title'].unique(),
    value='NBC',
    description='channel:',
    disabled=False,
)

def piegraph(ch):
  nbc = video[video['channel_title']==ch]
  lk = nbc['likes'].sum()
  dk = nbc['dislikes'].sum()
  cc = nbc['comment_count'].sum()
  plt.pie(x = [lk,dk,cc],labels = ['Like','Dislike','Comment'],autopct = '%1.1f%%')
  plt.show()
```

```python
widgets.interact(piegraph,ch =h )
```

```python
lr.predict([[40000,50,700,24]])
```

```python
lr.score(X_test,y_test)
```
